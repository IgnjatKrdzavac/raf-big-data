Na pocetku ucitavamo podatke koje zelimo da analiziramo.

```
path = 'dbfs:/projekat/'
csv_name = path + 'dfGames2.csv'

dfGames= spark.read.csv(csv_name, header=True, inferSchema=True)
dfGames.show(5, truncate=False)

csv_name1 = path + 'dfPlayers1.csv'

dfPlayers= spark.read.csv(csv_name1, header=True, inferSchema=True)
dfPlayers.show(5, truncate=False)

csv_name2 = path + 'dfAttendance1.csv'

dfAttendance= spark.read.csv(csv_name2, header=True, inferSchema=True)
dfAttendance.show(5, truncate=False)

```

Cuvamo dataframeove u obliku tabela nad kojima cemo moci da radimo SQL upite.

```
dfPlayers.createOrReplaceTempView("tablePlayers")
spark.sql("SELECT * from tablePlayers").show(5)
```

Analiza raspodele golova po pozicijama na terenu.

```
%sql

SELECT position, sum(goals) as goals
FROM tablePlayers
GROUP BY position 
ORDER BY goals desc
```

Prikaz svih igraca i broj minuta koji su odigrali na prvenstvu.

```
%sql

SELECT player, sum(minutes) as minutes
FROM tablePlayers
GROUP BY player 
ORDER BY minutes desc
```

Igraci sa najvise zutih kartona na prvenstvu.

```
%sql


SELECT player, cards_yellow
FROM tablePlayers
ORDER BY cards_yellow desc
```

Raspodela odigranih utakmica na prvenstvu, po timovima kojima pripadaju igraci.

```
%sql


SELECT club, sum(games) as games
FROM tablePlayers
GROUP BY club 
ORDER BY games desc
```

```
dfAttendance.createOrReplaceTempView("tableAttendance")
spark.sql("SELECT * from tableAttendance").show(5)
```

Prikaz najposecenijih stadiona na prvenstvu.

```
%sql
SELECT venue, max(attendance) as attendance
FROM tableAttendance
GROUP BY venue
ORDER BY attendance desc
```
